### Importing the Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Importing the Dataset

```python
dataset = pd.read_csv("./data/covid19_Confirmed_dataset.csv")
```

### Cleaning the Dataset

```python
df = dataset.drop(["Lat", "Long"], axis = 1, inplace = True)
```

```python
dataset.head()
```

### Aggregate the rows by country

```python
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

### Visualise data related to country

```python
corona_dataset_aggregated = corona_dataset_aggregated.apply(pd.to_numeric, errors="coerce")
```

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc["China"][:4].plot()
```

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

### Calculating Maximum Infection Rate of China, India, Spain

```python
## maximum infection rate 
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

### Calculating Maximum Infection of all countries

```python
countries = list(corona_dataset_aggregated.index)
max_infec_rates = []
for c in countries:
    max_infec_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Max_infection_rate"] = max_infec_rates
```

```python
corona_dataset_aggregated.head()
```

### Creating new dataframe of Maximum Infection Rate Only

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rate"])
```

```python
corona_data
```
